```python
#import your packages here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
import sqlalchemy
import seaborn as sns
```

```python
# Load the sales data here
dataset = pd.read_csv('sales_data.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().any()
```

```python
engine = sqlalchemy.create_engine('sqlite:///sales_data.db')
```

```python
dataset.to_sql('sales_table',engine)
```

```python
sales_data = pd.read_sql ('sales_table', engine)
```

```python
sales_data
```

**Question 1**

```python
#Question 1 i
de = '''SELECT Market, AVG(Sales) , AVG(Profit)
FROM sales_table
GROUP BY Market
ORDER BY AVG(Sales);'''

average_sales_and_profit = pd.read_sql (de, engine)
average_sales_and_profit
```

```python
#Question 1 ii

de = '''SELECT SUM(sales) AS Sales, SUM(Profit) AS Profit, Market
FROM sales_table
GROUP BY Market
ORDER BY Market'''

category_market = pd.read_sql (de, engine)
category_market
sns.set(style='white')
category_market.set_index('Market').plot(kind='bar', stacked=True, color=['steelblue', 'red'])
plt.ylabel('value')
plt.title('2013 Sales and Profit by Market', fontsize=16)
```

```python
#Question 1 iii
de ='''

SELECT Market, AVG(Sales) 
FROM sales_table
GROUP BY Market
ORDER BY AVG(Sales) DESC
LIMIT 1
'''
max_average_sales = pd.read_sql (de, engine)
max_average_sales
```

**The Market with the more sales on average is APAC**

**Question 2**

```python
#Question 2 i
de = '''SELECT Product_Name, SUM(Sales)
FROM sales_table
GROUP BY Product_Name
ORDER BY SUM(Sales) DESC
LIMIT 10;'''

best_Performing_products = pd.read_sql (de, engine)
best_Performing_products
```

```python
#Question 2 ii
de = '''SELECT Product_Name, SUM(Sales)
FROM sales_table
GROUP BY Product_Name
ORDER BY SUM(Sales)
LIMIT 10;'''

least_Performing_products = pd.read_sql (de, engine)
least_Performing_products
```
